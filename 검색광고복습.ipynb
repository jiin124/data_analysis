# 중점관리 키워드 추출
- 노출수, 클릭수 기준 상위 5%에 해당하는 키워드 추출
1. 95 백분위수 찾기 (quantile 함수 사용)
2. 95 백분위수 이상(상위 5%)의 노출수 추출
2. 상위 5%에 해당하는 키워드 추출

```python
imp=df['노출수']
imp
```

```python
imp.quantile()#기본 값은 2사분위 수 , 즉 중앙값
```

```python
imp.median()
```

```python
print(imp.quantile(0))#최솟값
print(imp.min())
```

```python
print(imp.quantile(1))#최댓값
print(imp.max())
```

```python
#95 백분위수 이상, 즉 5% 의 노출수 추출
imp=imp[imp>=imp.quantile(0.95)]
```

## 중점관리 키워드 선별
- 노출수와 클릭수 모두 95 백분위스 이상

## 조건이 여러개일 경우
- 데이터프레임[조건문&조건문]

```python
imp=df_index['노출수']
clk=df_index['클릭수']
```

```python
result=df_index[(imp>=imp.quantile(0.95))&(clk>=clk.quantile(0.95))]
```

```python
result
```
